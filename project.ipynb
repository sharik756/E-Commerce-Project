```python
x = pd.merge(x_order_report , x_sku_master ,  on = 'SKU')
x.head()
```

```python
x.rename(columns = { 'Weight (g)' : 'Weight_x'}, inplace = True)
x.rename(columns = { 'ExternOrderNo' : 'Order ID'}, inplace = True)
x['Weight_x'] = x['Weight_x'] /1000
def weight_slab_x(value):
    if value <= 0.5:
        return 0.5
    elif value > 0.5 and value <= 1:
        return 1
    elif value > 1 and value <= 1.5:
        return 1.5
x['weight_slab_x']  = x['Weight_x'].map(weight_slab_x)   
    
```
